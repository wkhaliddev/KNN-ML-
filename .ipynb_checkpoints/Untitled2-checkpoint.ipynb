```python
wr_max=ff.weighted_rating.max()
wr_min=ff.weighted_rating.min()
wr_min
```

```python
divider=wr_max-wr_min
norm_rating=[]
for i in range (ff.weighted_rating.size):
    norm=ff.weighted_rating[i]-wr_min
    norm/=divider
    norm_rating.append(norm)
```

```python
ff['norm_weighted_rating']=norm_rating
```
